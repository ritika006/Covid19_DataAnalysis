```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'], inplace= True, axis=1)
covid_df.head(10)
```

```python
## total no. of active cases = confirmed cases - deaths + cured

covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Deaths'] + covid_df['Cured'])
covid_df.head()
```

```python
# Distribution Statewise
statewise = pd.pivot_table(covid_df, values=['Confirmed','Deaths','Cured'], index = 'State/UnionTerritory', aggfunc = max)
statewise.head()
```

```python
# Recovery rate
statewise['Recovery_rate'] = (statewise['Cured']*100)/statewise['Confirmed']
```

```python
# Motality rate
statewise['Recovery_rate'] = (statewise['Deaths']*100)/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by='Confirmed', ascending = False)
```

```python
statewise.style.background_gradient(cmap = 'cubehelix')
```

```python
vaccination = vaccine_df.drop(
    columns = ['Sputnik V (Doses Administered)',
               'AEFI','18-44 Years (Doses Administered)',
                '45-60 Years (Doses Administered)',
                '60+ Years (Doses Administered)']
)
```

```python
vaccination.head()
```

```python
# plot pi chart between ratio of males and females vaccined

Males = vaccination['Male(Individuals Vaccinated)'].sum()
Females = vaccination['Female(Individuals Vaccinated)'].sum()
Transgender = vaccination['Transgender(Individuals Vaccinated)'].sum()

px.pie(
    names = ['Males','Females','Transgender'],
    values = [Males,Females,Transgender],
    title = 'Ratio of Males, Females and Transgender Vaccination'

)
```

```python
vaccination = vaccine_df[vaccine_df.State!='India']
vaccination.head()
```

```python
vaccination.rename(columns = {'Total Individuals Vaccinated' : 'Total'}, inplace = True)
```

```python
vaccination.head()
```

```python
# finding states with most number of vaccinated individuals

max_vac = vaccination.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total',ascending = False)[:5]
max_vac
```

```python
fig = pit.figure(figsize = (16,9))
pit.title('Top 5 vaccinaated states in India', size = 10)
x = sns.barplot(
    data = max_vac.iloc[:10],
    x = max_vac.index, y = max_vac.Total,
    linewidth=1, edgecolor = 'black',
    palette ='magma'
)

pit.xlabel('States')
pit.ylabel('Vaccination')
pit.show()
```
